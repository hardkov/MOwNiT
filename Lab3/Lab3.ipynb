## Zad 1 Przekszta≈Çcenie sfery w elipsoide
### 1 Sfera jednostkowa w 3d:

```python
def x(s, t):
    return np.cos(s) * np.sin(t)


def y(s, t):
    return np.sin(s) * np.sin(t)


def z(s, t):
    return np.cos(t)


s = np.linspace(0, 2 * np.pi, 25)
t = np.linspace(0, np.pi, 25)
ss, tt = np.meshgrid(s,t)

x_cords = x(ss,tt).flatten()
y_cords = y(ss,tt).flatten()    
z_cords = z(ss,tt).flatten()

fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot3D(x_cords, y_cords, z_cords)
plt.show()
```

### 2 Przekszta≈Çcenie w elipsioide:

```python
def print_ellipsoid(A, x, y, z):
    xyz = np.array([x, y, z]);
    e = A.dot(xyz)
    fig = plt.figure()
    ax = plt.axes(projection="3d")
    ax.plot3D(e[0], e[1], e[2])
    
    ellipsoid = np.row_stack([e[0], e[1], e[2]])
    plt.show()
    return ellipsoid

a = np.random.rand(3, 3)
b = np.random.rand(3, 3)
c = np.random.rand(3, 3)

ea = print_ellipsoid(a, x_cords, y_cords, z_cords)
eb = print_ellipsoid(b, x_cords, y_cords, z_cords)
ec = print_ellipsoid(c, x_cords, y_cords, z_cords)
```

### 3 Rozk≈Çad wed≈Çug warto≈õci osobliwych ka≈ºdej macierzy. Dodawanie wizualizacji p√≥≈Çosi:

```python
def draw_svd(u, s, e):
    x_s = y_s = z_s = [[0,0,0]] * 3
    x_e = y_e = z_e = [0,0,0]

    for i, x in enumerate(np.diag(s)):
        p = np.dot(u, x)
        
        x_e[i] = p[0]
        y_e[i] = p[1]
        z_e[i] = p[2]
        
    quiver1 = [x_s, y_s, z_s, x_e, y_e, z_e]


    fig = plt.figure()
    ax = fig.gca(projection='3d')
    ax.scatter(x_cords, y_cords, z_cords, marker=".")
    ax.scatter(e[0], e[1], e[2], color='purple', marker=".")
    ax.quiver(x_s, y_s, z_s, x_e, y_e, z_e, color='black')
    plt.show()

    
ua, sa, vha = np.linalg.svd(a);
ub, sb, vhb = np.linalg.svd(b);
uc, sc, vhc = np.linalg.svd(c);

draw_svd(ua, sa, ea)
draw_svd(ub, sb, eb)
draw_svd(uc, sc, ec)
```

### 4 Marcierz, kt√≥rej stosunek najwiƒôkszej oraz najmniejszej warto≈õci osobliwej jest wiƒôkszy od 100:

```python
d = np.random.rand(3,3)
u, s, vh = np.linalg.svd(d)
while max(s)/min(s) <= 100:
    d = np.random.rand(3,3)
    u, s, vh = np.linalg.svd(d)

out = print_ellipsoid(d, x_cords, y_cords, z_cords)
```

### 5 Wizualizacje $ SV_i^T, S\Sigma_iV_i^T\,oraz\,SU_i\Sigma_iV_i^T $ dla  macierzy  b :
#### a)  $ SV_i^T $

```python
out = print_ellipsoid(vhb.transpose(), x_cords, y_cords, z_cords)
```

Transforcmacja za pomocƒÖ macierzy V nie zmienia charakteru bry≈Çy, dalej pozostaje ona kulƒÖ. Wynika to z tego zƒô V jest unitarna.

#### b) $ S\Sigma_iV_i^T\ $

```python
xyz = np.array([x_cords, y_cords, z_cords]);

e = vhb.transpose().dot(np.diag(sb).dot(xyz))
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot3D(e[0], e[1], e[2])
plt.show()
```

Macierz sigma odpowiada za skalowanie. Otrzymujemy wiƒôc elipsoide

#### c) $ SU_i\Sigma_iV_i^T $

```python
e = vhb.transpose().dot(np.diag(sb).dot(ub.dot(xyz)))

fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot3D(e[0], e[1], e[2])
plt.show()
```

### 2 Obliczenie SVD macierzy pikseli I, a nastƒôpnie dokonanie przybli≈ºenia tej macierzy za pomocƒÖ low rank approximation:

```python
def low_rank_appr(filename, k):
    path, ext = os.path.splitext(filename)
    img = Image.open(filename)
    img = img.convert("L")
    img = np.array(list(img.getdata(band=0)), float)
    img.shape = (512, 512)
    img_arr = np.asarray(img)

    u, s, vh = np.linalg.svd(img_arr)

    new_img_arr = np.matrix(u[:, :k]) * np.diag(s[:k]) * np.matrix(vh[:k, :])
    file = path + '_' + str(k) + ext
    new_img = Image.fromarray(np.uint8(new_img_arr))
    new_img.save(file)
```

### 3  Por√≥wnanie obrazu wynikowego z obrazem ≈∫r√≥d≈Çowym dla r√≥≈ºnych warto≈õci k

```python
list_of_apprs = [1, 2, 5, 15, 25, 55, 85, 100]

for i in list_of_apprs:
    low_rank_appr("pepe.jpg", i)
```
